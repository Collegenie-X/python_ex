- Barplot: 매출, 고객 유형, 요일별 분석
- Boxplot: 고객 유형, 지출 금액 분포 분석
- Scatterplot: 두 변수 간의 관계 분석 (매출과 팁)
- Lineplot: 월별 트렌드 분석
- Heatmap: 카테고리 간 상관 분석
- Violinplot: 연령대별 소비 패턴
- Pairplot: 다변량 분석


### 1. 요일별 매출 분석 (Barplot)
- 목표: 요일별 평균 매출을 분석하여, 특정 요일에 판매가 집중되는지 파악합니다.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import random

# 100개 랜덤 데이터 생성: 요일별 매출
data = {
    "day": [
        random.choice(["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"])
        for _ in range(100)
    ],
    "sales": [
        random.randint(100, 500) for _ in range(100)
    ],  # 100~500 사이의 랜덤 매출 금액
}

# 데이터프레임 생성
df = pd.DataFrame(data)

# 요일별 평균 매출 시각화
sns.barplot(x="day", y="sales", data=df)
plt.title("Average Sales by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Sales ($)")
plt.show()
```

### 2. 고객 유형별 지출 분석 (Boxplot)
- 목표: 고객 유형별 지출의 분포를 분석하여, 고객 세그먼트가 매출에 미치는 영향을 파악합니다.

```python
# 100개 랜덤 데이터 생성: 고객 유형별 지출 금액
data = {
    "customer_type": [random.choice(["Regular", "New"]) for _ in range(100)],
    "spending": [
        random.randint(50, 300) for _ in range(100)
    ],  # 50~300 사이의 랜덤 지출 금액
}

# 데이터프레임 생성
df = pd.DataFrame(data)

# 고객 유형별 지출 분포 분석
sns.boxplot(x="customer_type", y="spending", data=df)
plt.title("Spending Distribution by Customer Type")
plt.xlabel("Customer Type")
plt.ylabel("Spending ($)")
plt.show()
```

### 3. 판매 금액과 팁 간의 관계 분석 (Scatterplot)
- 목표: 판매 금액과 팁 금액 간의 상관 관계를 분석하여, 팁 금액이 판매 금액에 따라 어떻게 변화하는지 확인합니다.

```python
# 100개 랜덤 데이터 생성: 판매 금액과 팁 금액
data = {
    "total_bill": [
        random.randint(30, 100) for _ in range(100)
    ],  # 30~100 사이의 랜덤 판매 금액
    "tip": [
        random.choice([1, 2, 3, 4, 5, 6, 7, 8]) for _ in range(100)
    ],  # 팁 금액 (1~8 사이의 랜덤 팁)
}

# 데이터프레임 생성
df = pd.DataFrame(data)

# 판매 금액과 팁 간의 관계 분석
sns.scatterplot(x="total_bill", y="tip", data=df)
plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

## 4. 월별 매출 변화 분석 (Lineplot)
- 목표: 월별 매출 추이를 시각화하여, 계절적 변화나 판매 트렌드를 파악합니다.

```python
# 100개 랜덤 데이터 생성: 월별 매출
months = [
    "Jan",
    "Feb",
    "Mar",
    "Apr",
    "May",
    "Jun",
    "Jul",
    "Aug",
    "Sep",
    "Oct",
    "Nov",
    "Dec",
]
data = {
    "month": [random.choice(months) for _ in range(1000)],
    "sales": [
        random.randint(200, 500) for _ in range(1000)
    ],  # 200~500 사이의 랜덤 매출 금액
}

# 데이터프레임 생성
df = pd.DataFrame(data)

# 월별 매출 추이 시각화
sns.lineplot(x="month", y="sales", data=df, estimator="mean", marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Average Sales ($)")
plt.show()
```

## 5. 고객 연령대별 구매 패턴 분석 (Violinplot)
- 목표: 연령대별 소비 패턴을 분석하여, 특정 연령대의 소비 성향을 파악합니다.

```python
# 100개 랜덤 데이터 생성: 고객 연령대별 구매 금액
count = 30

data = {
    "age_group": [
        random.choice(["18-25", "26-35", "36-45", "46-55", "56+"]) 
        for _ in range(count)
    ],
    "purchase_amount": [
        random.randint(30, 200) for _ in range(count)
    ],  # 30~200 사이의 랜덤 구매 금액
}

# 데이터프레임 생성
df = pd.DataFrame(data)

# 고객 연령대별 구매 패턴 분석
sns.violinplot(x="age_group", y="purchase_amount", data=df)
plt.title("Purchase Amount by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Purchase Amount ($)")
plt.show()
```

## 6. 제품 카테고리별 판매 성과 분석 (Heatmap)
- 목표: 제품 카테고리 간 상관 관계를 분석하여, 각 카테고리의 판매 성과를 비교합니다.

```python
import numpy as np

# 100개 랜덤 데이터 생성: 카테고리별 판매 성과 (매출)
data = np.random.randint(100, 500, size=(5, 5))  # 5x5 행렬로 랜덤 매출 데이터 생성
categories = ["Electronics", "Clothing", "Furniture", "Groceries", "Toys"]

# 히트맵 생성
sns.heatmap(
    data, annot=True, cmap="YlGnBu", xticklabels=categories, yticklabels=categories
)
plt.title("Sales Performance by Category")
plt.show()
```

### 7. 지출과 수입의 관계 분석 (Pairplot)
- 목표: 지출과 수입 간의 관계를 분석하여, 두 변수 간의 상관성을 파악합니다.

```python
# 100개 랜덤 데이터 생성: 지출과 수입 데이터
data = {
    "income": [
        random.randint(2500, 5000) for _ in range(100)
    ],  # 2500~5000 사이의 랜덤 수입
    "spending": [
        random.randint(1500, 3500) for _ in range(100)
    ],  # 1500~3500 사이의 랜덤 지출
    "savings": [
        random.randint(500, 2000) for _ in range(100)
    ],  # 500~2000 사이의 랜덤 저축
}
```

```python
# 예시 데이터: 지출과 수입 데이터
data = {
    "income": [2500, 3000, 3500, 4000, 4500],
    "spending": [1500, 1700, 1800, 1900, 2100],
    "savings": [1000, 1300, 1400, 1500, 1200],
}
```

```python
# 데이터프레임 생성
df = pd.DataFrame(data)

# 지출과 수입 간 관계 분석
sns.pairplot(df)
plt.suptitle("Income, Spending, and Savings Relationships", y=1.02)
plt.show()
```
