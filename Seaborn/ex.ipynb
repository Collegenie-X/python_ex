```python
import seaborn as sns
import matplotlib.pyplot as plt

# seaborn 기본 데이터셋 불러오기
tips = sns.load_dataset('tips')

# 간단한 히스토그램
sns.histplot(tips['total_bill'], kde=True)
plt.title('Total Bill Distribution')
plt.show()
```

## 목표: **히스토그램(histplot)**과 **밀도 추정(displot)**을 사용하여 데이터의 분포를 시각화한다.

- histplot: 데이터를 구간별로 나누어 분포를 시각화
- displot: 히스토그램과 커널 밀도 추정(KDE)을 함께 시각화

```python
# 매출 추이 분석
sns.displot(tips['total_bill'], kde=True, bins=30)
plt.title('Distribution of Total Bill')
plt.show()
```

### 목표: **바 플롯(barplot)**과 **카운트 플롯(countplot)**을 사용하여 카테고리별 평균/빈도 분석을 시각화한다.

- barplot: 카테고리 변수에 대한 평균값을 계산하여 시각화
- countplot: 각 카테고리의 개수를 세서 시각화

```python
# 요일별 평균 매출
sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Average Total Bill by Day')
plt.show()

# 요일별 주문 수
sns.countplot(x='day', data=tips)
plt.title('Order Count by Day')
plt.show()
```


### 목표: **박스 플롯(boxplot)**과 **바이올린 플롯(violinplot)**을 사용하여 데이터의 분포와 이상치를 분석한다.

- boxplot: 데이터의 사분위수 및 이상치를 시각화
- violinplot: 데이터의 밀도와 분포를 동시에 보여주는 플롯

```python
# Box Plot: 요일별 매출 분포
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Distribution of Total Bill by Day')
plt.show()

# Violin Plot: 요일별 매출 밀도
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('Density of Total Bill by Day')
plt.show()
```

## **라인 플롯(lineplot)**과 **포인트 플롯(pointplot)**을 사용하여 시간 추세와 두 변수 간의 관계를 시각화한다.

- lineplot: 시간에 따른 추이 분석
- pointplot: 그룹별 평균을 포인트로 시각화

```python
# Line Plot: 시간대별 매출 추이
sns.lineplot(x='time', y='total_bill', data=tips)
plt.title('Total Bill Trend by Time')
plt.show()

# Point Plot: 요일별 평균 매출 변화
sns.pointplot(x='day', y='total_bill', data=tips)
plt.title('Average Total Bill by Day')
plt.show()
```

## 목표: **산점도(scatterplot)**와 **페어 플롯(pairplot)**을 사용하여 변수 간 관계를 분석한다.

- scatterplot: 두 연속형 변수 간의 관계 시각화
- pairplot: 여러 변수 간의 관계를 한 번에 시각화

```python
# Scatter Plot: 매출과 팁 간 관계
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Relationship between Total Bill and Tip')
plt.show()

# Pair Plot: 여러 변수 간의 관계 분석
sns.pairplot(tips)
plt.show()
```

## 목표: **히트맵(heatmap)**을 사용하여 상관관계 분석을 시각화한다.

- heatmap: 변수 간의 상관 관계를 시각적으로 표현

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
tips = sns.load_dataset("tips")

# Filter out non-numeric columns
numeric_tips = tips.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = numeric_tips.corr()

# Create the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Tips Dataset')
plt.show()
```
