```python
import numpy as np
import pandas as pd
# seaborn
import seaborn as sns
COLORS = sns.color_palette()

%matplotlib inline
```

#### 데이터 로딩
1. 정제된 데이터 사용(retail.csv)

```python
dtypes = {
    'UnitPrice': np.float32,
    'CustomerID': np.int32,
    'Quantity': np.int32
}
retail = pd.read_csv('../source_data/OnlineRetailClean.csv', dtype=dtypes)
retail.head()
```

#### 날짜 타입 데이터 변환
 - 문자열로 로딩하는 것보다 date/datetime 타입으로 로딩하는 것이 분석에 용이

```python
retail['InvoiceDate'] = pd.to_datetime(retail['InvoiceDate'], infer_datetime_format=True)
retail.info()
```

#### 해당 기간 동안의 매출
 - 전체 매출
 - 국가별 매출
 - 월별 매출
 - 요일별 매출
 - 시간별 매출

#### 전체 매출

```python
total_revenue = retail['CheckoutPrice'].sum()
print(f'{total_revenue:,}')
```

#### 국가별 매출

```python
rev_by_countries = retail.groupby('Country').agg({'CheckoutPrice': 'sum'}).sort_values(by='CheckoutPrice')
rev_by_countries
```

```python
print(round(rev_by_countries / total_revenue,5) *100)
```

```python
plot = rev_by_countries.plot(kind='bar', color=COLORS[-1], figsize=(20, 10))
plot.set_xlabel('Country', fontsize=11)
plot.set_ylabel('Revenue (%)', fontsize=11)
plot.set_title('Revenue by Country', fontsize=13)
plot.set_xticklabels(labels=rev_by_countries.index, rotation=45)
```

#### 그래프 유틸 함수

```python
def plot_bar(df, xlabel, ylabel, title, color=COLORS[0], figsize=(20, 10), rotation=45):
    plot = df.plot(kind='bar', color=color, figsize=figsize)
    plot.set_xlabel(xlabel, fontsize=11)
    plot.set_ylabel(ylabel, fontsize=11)
    plot.set_title(title, fontsize=13)
    plot.set_xticklabels(labels=df.index, rotation=rotation)
                   
plot_bar(rev_by_countries, 'Country', 'Revenue(%)', 'Revenue by Country')
```

#### 월별 매출

```python
retail['InvoiceDate'].sort_values(ascending=False)
```

```python
def extract_month(date):
    month = str(date.month)
    if date.month < 10:
        month = '0' + month
    return str(date.year) + month 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 월별 매출 합계 계산 (retail 데이터프레임을 사용)
rev_by_month = retail.set_index('InvoiceDate').groupby(extract_month).sum()['CheckoutPrice']

# 월별 매출 데이터를 데이터프레임으로 변환
rev_by_month_df = rev_by_month.reset_index()

# Seaborn을 사용하여 막대 그래프 그리기
plt.figure(figsize=(12, 6))
sns.barplot(x=rev_by_month_df['InvoiceDate'], y=rev_by_month_df['CheckoutPrice'], palette='Blues_d')

# 그래프 제목과 레이블 설정
plt.xlabel('Month', fontsize=11)
plt.ylabel('Revenue', fontsize=11)
plt.title('Revenue by Month', fontsize=13)

# x축 레이블 회전
plt.xticks(rotation=45)

# 그래프 출력
plt.show()
```

#### top 3 아이템의 월별 판매량 추이

```python
monthly_top3 = retail.set_index('InvoiceDate').groupby(['StockCode', extract_month]).sum()[['Quantity', 'CheckoutPrice']].loc[top_selling.index]
monthly_top3
```

```python
monthly_top3['CheckoutPrice']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 'InvoiceDate'에서 월만 추출하여 'Month' 열 생성
monthly_top3['Month'] = monthly_top3.index.get_level_values('InvoiceDate').astype(str).str[:6]  # 예: '201101'

# Seaborn을 사용하여 'Product'와 'Month'별 매출 시각화
plt.figure(figsize=(14, 8))

# 'Product'와 'Month'를 결합하여 x축에 표시할 수 있도록 준비
monthly_top3['Product/Month'] = monthly_top3.index.get_level_values('StockCode').astype(str) + ' / ' + monthly_top3['Month']

# Seaborn 바 차트
sns.barplot(x='Product/Month', y='CheckoutPrice', data=monthly_top3.reset_index(), palette='Blues_d')

# 그래프 제목과 레이블 설정
plt.xlabel('Product / Month', fontsize=11)
plt.ylabel('Revenue', fontsize=11)
plt.title('Revenue of Top 3 Items by Product and Month', fontsize=13)

# x축 레이블 회전
plt.xticks(rotation=45)

# 그래프 출력
plt.show()
```
