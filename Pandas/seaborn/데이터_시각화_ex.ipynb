- 특징	Matplotlib	Seaborn
- 시각화 수준	저수준 (Low-level)	고수준 (High-level)
- 사용 용도	복잡한 커스터마이징이 필요한 시각화	복잡한 시각화를 쉽게 작성, 통계적 시각화에 특화
- 통계적 분석 기능	없음 (수동으로 회귀선 추가 등 해야 함)	내장된 회귀선, 분포 분석, 상관 관계 분석 등 제공
- 스타일 및 디자인	사용자 정의가 가능하지만 코드가 복잡함	다양한 미리 설정된 스타일과 색상 팔레트 제공
- pandas 통합	pandas 데이터프레임과 통합할 수 있지만 수동으로 처리해야 함	pandas 데이터프레임과 기본적으로 잘 통합되어 있음

```python
import seaborn as sns
import matplotlib.pyplot as plt

# seaborn 기본 데이터셋 불러오기
tips = sns.load_dataset('tips')

# 간단한 히스토그램
sns.histplot(tips['total_bill'], kde=True)
plt.title('Total Bill Distribution')
plt.show()
```

### 1. 히스토그램 (Histplot) - 데이터 분포 분석
- 사용 경우: 특정 변수의 분포를 파악하거나, 데이터가 특정 구간에 몰려 있는지 확인할 때 사용합니다. 예를 들어, 고객의 구매 금액 분포를 분석하거나, 팁 금액의 분포를 확인하는 데 유용합니다.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 예시 데이터 (pandas DataFrame)
data = {
    'total_bill': [23.68, 25.34, 17.92, 21.56, 30.88, 13.91, 22.85, 19.42, 32.68, 27.45],
    'tip': [3.31, 4.08, 3.06, 3.71, 4.56, 2.01, 3.63, 3.08, 4.91, 3.93]
}
df = pd.DataFrame(data)

# 히스토그램을 사용해 total_bill(판매 금액)의 분포를 시각화
sns.histplot(df['total_bill'], kde=True, bins=10)
plt.title('Total Bill Distribution')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

## 2. 바 플롯 (Barplot) - 카테고리별 평균값 분석
- 사용 경우: 카테고리 변수별로 데이터를 집계하여 평균값을 비교할 때 사용합니다. 예를 들어, 요일별 평균 매출을 비교하거나 성별 팁 차이를 확인할 때 사용합니다.
- 실전 예시: 요일별 판매 금액 비교

```python
# 예시 데이터 (pandas DataFrame)
data = {
    'day': ['Sun', 'Sun', 'Mon', 'Mon', 'Tue', 'Tue', 'Wed', 'Wed'],
    'total_bill': [23.68, 25.34, 17.92, 21.56, 30.88, 13.91, 22.85, 19.42]
}
df = pd.DataFrame(data)

# 바 플롯을 사용해 요일별 평균 total_bill(판매 금액)을 시각화
sns.barplot(x='day', y='total_bill', data=df)
plt.title('Average Total Bill by Day')
plt.show()
```

### 3. 박스 플롯 (Boxplot) - 데이터 분포 및 이상치 확인
- 사용 경우: 데이터의 분포를 확인하고 이상치를 파악할 때 사용합니다. 예를 들어, 각 직원별 팁의 분포를 확인하거나 요일별 매출의 이상치를 점검할 때 사용됩니다.
- 실전 예시: 성별 팁 금액 분포

```python
# 예시 데이터 (pandas DataFrame)
data = {
    'sex': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'tip': [3.31, 4.08, 3.06, 3.71, 4.56, 2.01, 3.63, 3.08]
}
df = pd.DataFrame(data)

# 박스 플롯을 사용해 성별로 팁 금액의 분포 확인
sns.boxplot(x='sex', y='tip', data=df)
plt.title('Tip Distribution by Gender')
plt.show()
```

### 4. 라인 플롯 (Lineplot) - 시간 추세 분석
- 사용 경우: 시간에 따른 추세를 분석하고, 특정 변수의 변화를 추적할 때 사용합니다. 예를 들어, 주간 매출 추이나 고객 수 변화를 시각화하는 데 유용합니다.

```python
# 예시 데이터 (pandas DataFrame)
data = {
    'time': ['Dinner', 'Dinner', 'Lunch', 'Lunch', 'Dinner', 'Lunch', 'Dinner', 'Lunch'],
    'total_bill': [23.68, 25.34, 17.92, 21.56, 30.88, 13.91, 22.85, 19.42]
}
df = pd.DataFrame(data)

# 라인 플롯을 사용해 시간대별 매출 추이 시각화
sns.lineplot(x='time', y='total_bill', data=df)
plt.title('Total Bill Trend by Time')
plt.show()
```

### 5. 히트맵 (Heatmap) - 상관 관계 분석
- 사용 경우: 여러 변수 간의 상관 관계를 시각적으로 분석할 때 사용됩니다. 예를 들어, 매출과 팁 간 상관 관계를 분석하거나, 여러 변수 간의 관계를 시각화하는 데 유용합니다.
- 실전 예시: 매출과 팁 간의 상관 관계 분석

```python
# 예시 데이터 (pandas DataFrame)
data = {
    'total_bill': [23.68, 25.34, 17.92, 21.56, 30.88],
    'tip': [3.31, 4.08, 3.06, 3.71, 4.56]
}
df = pd.DataFrame(data)

# 상관 관계 계산
correlation_matrix = df.corr()

# 히트맵을 사용해 상관 관계 시각화
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
